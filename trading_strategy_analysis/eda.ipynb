```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv", index_col=0)
df.head()
```

```python
print(f"Bh Mean: {df["bh_cagrs"].mean()}")
print(f"SMA Mean: {df["sma_cagrs"].mean()}")
```

```python
print(f"Bh Meadian: {df["bh_cagrs"].median()}")
print(f"SMA Meadian: {df["sma_cagrs"].median()}")
```

```python
print(f"Bh Std: {df["bh_cagrs"].std()}")
print(f"SMA Std: {df["sma_cagrs"].std()}")
```

```python
print(f"Bh Max: {df["bh_cagrs"].max()}")
print(f"Sma Max: {df["sma_cagrs"].max()}")
```

```python
print(f"Bh Min: {df["bh_cagrs"].min()}")
print(f"Sma Min: {df["sma_cagrs"].min()}")
```

```python
print(f"SMA Strategies that had two total trades {len(df[df["sma_num_trades "] == 2])} -> {len(df[df["sma_num_trades "] == 2]) / len(df) * 100}%")
print(f"SMA Strategies that had 0 total trades {len(df[df["sma_num_trades "] == 0])} -> {len(df[df["sma_num_trades "] == 0]) / len(df) * 100}%")
print(f"SMA Strategies that had move than 10 total trades {len(df[df["sma_num_trades "] > 10])} -> {len(df[df["sma_num_trades "] > 10]) / len(df) * 100}%")
```

```python
print(f"Bh Mean len: {df["bh_lens"].mean()}")
print(f"SMA Mean len: {df["sma_lens"].mean()}")

print(f"Bh Median len: {df["bh_lens"].median()}")
print(f"SMA Median len: {df["sma_lens"].median()}")

print(f"Bh Min len: {df["bh_lens"].min()}")
print(f"SMA Min len: {df["sma_lens"].min()}")

print(f"Bh Max len: {df["bh_lens"].max()}")
print(f"SMA Max len: {df["sma_lens"].max()}")
```
